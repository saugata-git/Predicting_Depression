# Creating the data frame

```python
def process_results(results):
    id_list = [status.id for status in results]
    data_set = pd.DataFrame(id_list, columns=["id"])
    
    # Processing Tweet Data

    data_set["text"] = [status.text for status in results]
    data_set["created_at"] = [status.created_at for status in results]
    data_set["retweet_count"] = [status.retweet_count for status in results]
    data_set["like"] = [status.favorite_count for status in results]
    data_set["source"] = [status.source for status in results]
   
    # Processing User Data
    data_set["user_id"] = [status.author.id for status in results]
    data_set["user_screen_name"] = [status.author.screen_name for status in results]
    data_set["user_name"] = [status.author.name for status in results]
    data_set["user_created_at"] = [status.author.created_at for status in results]
    data_set["user_description"] = [status.author.description for status in results]
    data_set["user_followers_count"] = [status.author.followers_count for status in results]
    data_set["user_friends_count"] = [status.author.friends_count for status in results]
    data_set["user_location"] = [status.author.location for status in results]
    data_set["date"] = data_set["created_at"].dt.date
    data_set["time"] = data_set["created_at"].dt.time
    
    return data_set
data_set = process_results(results)
```

```python
import difflib
def fraction_of_question_centric_posts(screen_name):
    results=[]
    tweets = api.user_timeline(screen_name , count = 100, include_rts = True)
    for i in range(0,len(tweets)):
        results.append(tweets[i])  
    id_list = [status.id for status in results]
    data_set5 = pd.DataFrame(id_list, columns=["id"])    
    data_set5["created_at"] = [status.created_at for status in results]
    data_set5["date"] = data_set5["created_at"].dt.date
    data_set5["time"] = data_set5["created_at"].dt.time
    data_set5["text"] = [status.text for status in results]
   
    listsQ=[]

    Sample_Questions = ["what is the weather like","where are we today","why did you do that","where is the dog","when are we going to leave","why do you hate me","what is the Answer to question 8",
                    "what is a dinosour","what do i do in an hour","why do we have to leave at 6.00", "When is the apointment","where did you go","why did you do that","how did he win","why wont you help me",
                    "when did he find you","how do you get it","who does all the shipping","where do you buy stuff","why dont you just find it in the target","why don't you buy stuff at target","where did you say it was",
                    "when did he grab the phone","what happened at seven am","did you take my phone","do you like me","do you know what happened yesterday","did it break when it dropped","does it hurt everyday",
                    "does the car break down often","can you drive me home","where did you find me","Have you brushed your teeth?","Does he like to read?","Can they come downstairs?","Would you like this book or that book?",
                    "Should I get coffee, tea, or water?","You have a new sweater, right?","The snow is pretty, isnt it?","how much time ?",
                    "can it fly from here to target","could you find it for me","is not it?"," Does he did it","who  ?","is he/she     ?","  Am i the stuff ?","  ?"]


    for text in data_set5["text"]:
                for Ran_Question in Sample_Questions:
                  
                    Question_Matcher = difflib.SequenceMatcher(None,Ran_Question ,text)
                    QM=  Question_Matcher.ratio() 
                    if QM>0.5:
                        listsQ.append(1)
                        break
                else:    
                     listsQ.append(0)
                        
    se = pd.Series(listsQ)
    data_set5['question_centric'] = se.values                  

    print( data_set5.groupby("date")["question_centric"].sum())
    
```
